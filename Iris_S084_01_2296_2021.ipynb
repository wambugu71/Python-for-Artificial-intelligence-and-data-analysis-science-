```python
import pandas as pd
from sklearn.datasets import load_iris

#1. Load data into dataframe and print shape, dtypes, and first 7 rows
df = pd.read_csv('/content/IRIS.csv')
print(df.shape)
print(df.dtypes)
print(df.head(7))

#2. Print keys, shape, feature names, and description
data = load_iris()
print(data.keys())
print(data.data.shape)
print(data.feature_names)
print(data.DESCR)

#3a. Number of observations
print(df.info())
print(len(df))
print(len(df.columns))
print(len(df.index))
print(df.shape)
print(df.shape[0])
print(df.shape[1])
print(df.size)
print(df['sepal_length'].count())

#3b. Number of missing values
print(df.isnull())
print(df.isnull().sum())

#4. Number of columns and total missing values
print(df.isnull().sum()[df.isnull().sum() > 0].count())
print(df.isnull().sum().sum())

#5. Dealing with missing values
df.fillna(0) #replace with 0
df.fillna(method='ffill') #fill with next value
df.fillna(method='bfill') #fill with previous value
df.interpolate() #interpolate
df.dropna() #drop rows

#6. Convert categorical to numerical
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['species'] = le.fit_transform(df['species'])

#7. View statistical details
print(df.describe())

#8. Drop id columns
#df.drop(columns=['id'], inplace=True)
```
